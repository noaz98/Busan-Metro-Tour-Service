```python

# 1호선 관광지 딕셔너리
first_station = {'다대포해수욕장': '다대포 해수욕장', '다대포항': None, '낫개': None, 
                 '신장림': None, '장림': None, '동매': None,
                 '신평': None, '하단': '가덕도 해안산책로', '당리': None, '사하': None, 
                 '괴정': None, '대티': None, '서대신': None, '동대신': None ,'닥바': None,
                  '토성':'감천 문화마을', '자갈치':'자갈치 시장', '남포':'깡통 시장',
                  '중앙': '영도 대교', '부산역': '차이나 타운', '초량': None, '부산진': None,
                  '좌천': None, '범일': None, '범내골': None, '서면(환승역)': None, '부전':'부산 시민공원',
                  '양정': None, '시청': None, '연산': None, '교대': None, '동래':'동래 온천',
                  '명륜': None, '온천장': '온천천', '부산대': '부산대학교', '장전': None,
                  '구서': None, '두실': None, '남산': None, '범어사': None, '노포': None}
# 2호선 관광지 딕셔너리
second_station = {'양산': None, '남양산': None, '부산대양산캠퍼스': None, '증산': None,
                  '호포': None, '금곡': None, '동원': None, '율리': None, '화명': None,
                  '수정': '낙동강 생태공원', '덕천': None, '구명': None, '구남': None,
                  '모라': None, '모덕': None, '덕포': None, '사상': '삼락 생태공원', '감전': None,
                  '주례': None, '냉정': None, '개금': None, '동의대': None, '가야': None, '부암': None,
                  '서면(환승역)': None, '전포': '전포 카페거리', '국제금융센터': None, '문현': None,
                  '지게골': None, '못골': None, '대연': 'UN 기념공원', '경성대-부경대': None,
                  '남천': None, '광안': '광안리 해수욕장', '수영': None, '민락': None,
                  '센텀시티': '신세계 백화점', '벡스코': None, '동백': '시립 미술관', '해운대': '해운대 해수욕장',
                  '중동': None, '장산': '송정 해수욕장'}

# 딕셔너리로 저장한 역들을 새로운 변수에 리스트로 저장
# 정방향 1,2호선
f_data = [k for k in first_station.keys()]
s_data = [k for k in second_station.keys()]

# 서면을 기준으로 좌/우 구분
a = list(reversed(s_data))
sec1_staion = s_data[25:]
sec2_station = a[18:]

# value를 이용한 key 찾기
def find_values(station,value_input):
    find_k = [k for k, v in station.items() if v == value_input]
    if find_k == []:
        return 
    else:
        # 리스트 인덱스만 추출
        return ''.join(find_k) 
    
# input 받은 값이 station 안에 있을 때 그 위치 전까지 새로운 배열 생성 + reverse 
def new_list(station,start,end):
    new = []
    for i in station:
        if i == start:
            new = station[len(new):]
            if end in new:
                return new
            else:
                station = list(reversed(station))
                start_index = station.index(start)
                end_index = station.index(end)
                new = station[start_index:end_index+1]
                return new
        new.append(i)
    return new
```

```python
class Node():
    def __init__(self):
        self.data = None
        self.link = None
        self.left_link = None
        self.right_link = None

def printNodes(start,end):
    current = start
    if current == None:
        return 
    print(current.data,end=' -> ')
    while current.link != None:
        current = current.link
        if current.data != end:
            print(current.data,end=' -> ')
        else:
            print(current.data)
    print(end)
    
def node_station(station,data1,data2,end):    
    head,pre = None,None
    memory = []
    
    node = Node()
    node.data = station[0]
    head = node
    memory.append(node)
    
    if end in station:
        for data in station[1:]: 
            if data == end :
                break
            pre = node
            node = Node()
            node.data = data
            pre.link = node
            memory.append(node)
            
    elif end in data1 or end in data2:
        if end in data1:
            a = station.index('서면(환승역)')
            for data in station[1:a+1]:
                pre = node
                node = Node()
                node.data = data
                pre.link = node
                memory.append(node)
            for data in data1[0:]:
                if data == end:
                    break
                pre = node
                node = Node()
                node.data = data
                pre.link = node
                memory.append(node)
        else:
            a = station.index('서면(환승역)')
            for data in station[1:a+1]:
                pre = node
                node = Node()
                node.data = data
                pre.link = node
                memory.append(node)
            for data in data2[0:]:
                if data == end:
                    break
                pre = node
                node = Node()
                node.data = data
                pre.link = node
                memory.append(node)
                
    printNodes(head,end)
```
